1. Что такое искусственная нейронная сеть (ИНС)?
Искусственная нейронная сеть - это система нейронов, которые взаимодействуют между собой. Каждый нейрон принимает сигналы или же отправляет их другим процессорам (нейронам). Объединённые в одну большую сеть, нейроны, обучаясь, могут выполнять сложные задачи.
Нейрон – базовая единица нейронной сети. У каждого нейрона есть определённое количество входов, куда поступают сигналы, которые суммируются с учётом значимости (веса) каждого входа. Далее сигналы поступают на входы других нейронов. Вес каждого такого «узла» может быть как положительным, так и отрицательным. Например, если у нейрона есть четыре входа, то у него есть и четыре весовых значения, которые можно регулировать независимо друг от друга.

6. Объясните понятие «обучение с учителем» и «обучение без учителя».
Обучение с учителем (supervised learning) — это алгоритм обучения, для которого нужны данные, на основе которых будет строиться модель. К supervised learning относятся задачи классификации, прогнозирования, ранжирования и регрессии.

Обучение без учителя (unsupervised learning) — это алгоритм обучения, для которого задействуются неклассифицированные данные. В этих данных нужно найти закономерности, а также создать их структуру. К формам обучения без учителя относят, например, кластеризацию — разбивку данных на похожие категории; уменьшение размерности — снижение количества признаков наборов данных.

8. Расскажите о методе обратного распространения ошибки (backpropagation).
По сути, обратное распространение — это алгоритм, используемый для быстрого вычисления производных. Искусственные нейронные сети используют обратное распространение в качестве алгоритма обучения для вычисления градиентного спуска относительно весов. Желаемые выходные данные сравниваются с достигнутыми выходными данными системы, а затем системы настраиваются путем корректировки весов соединений, чтобы максимально уменьшить разницу между ними. Алгоритм получил свое название, потому что веса обновляются в обратном направлении, от вывода к вводу.
9. В чем заключается метод градиентного спуска и как он применяется в обучении ИНС?
Метод градиентного спуска — это один из методов оптимизации, который позволяет нейронной сети обучаться.

Он помогает находить направление, в котором функция ошибки уменьшается, и обновлять параметры модели соответствующим образом.

Алгоритм градиентного спуска включает следующие шаги:

Инициализация параметров модели случайными значениями.

Подача входных данных в модель и получение предсказаний.

Вычисление значения функции ошибки, сравнение предсказаний с фактическими значениями.

Определение градиента функции ошибки по каждому параметру модели.

Обновление значений параметров по формуле: новое значение параметра = старое значение параметра − learning rate * градиент.

Повторение шагов 2–5 для каждого этапа прогонки обучения (эпохи) или до достижения критерия остановки.

Цель алгоритма градиентного спуска — минимизировать критерий качества нейронной сети — сумму квадратов ошибки обучающей выборки.
10. Что такое недообучение(underfitting) и как его можно избежать?
Недообучение — ситуация, когда в параметрическом семействе функций не удается найти функцию, которая хорошо описывает данные.
Самая частая причина недообучения — когда сложность устройства данных выше, чем сложность устройства модели, которую придумал исследователь. Решением для такой проблемы будет усложнение модели и поиск лучшего описания эффектов, которые есть в данных.
11. Что такое переобучение (overfitting) и как его можно избежать?
Переобучение (overfitting) — одна из проблем глубоких нейронных сетей (Deep Neural Networks, DNN), состоящая в следующем: модель хорошо объясняет только примеры из обучающей выборки, адаптируясь к обучающим примерам, вместо того чтобы учиться классифицировать примеры, не участвовавшие в обучении (теряя способность к обобщению).Это проблема недостаточных наборов данных и лечится она увеличением количества обучающих примеров.
Возможные действия при переобучении:
Увеличение количества данных в наборе;
Уменьшение количества параметров модели;
Добавление регуляризации / увеличение коэффициента регуляризации. Возможные действия при недообучении:
Добавление новых параметров модели;
Использование для описания модели функций с более высокой степенью;
Уменьшение коэффициента регуляризации.
12. Объясните понятие регуляризации и приведите примеры методов регуляризации.
Регуляризация в статистике, машинном обучении, теории обратных задач — метод добавления некоторых дополнительных ограничений к условию с целью решить некорректно поставленную задачу или предотвратить переобучение. Эта информация часто имеет вид штрафа за сложность модели. Например, это могут быть ограничения гладкости результирующей функции или ограничения по норме векторного пространства.
В машинном обучении существуют следующие методы регуляризации:
Регуляризация по технике «Лассо» (L1). Штрафует весовые коэффициенты модели для изменения их важности для модели. Может привести к полному обнулению переменных. Применяется, когда набор данных включает в себя много переменных и необходимо исключить некоторые из них, чтобы лучше понять, как важные признаки влияют на модель.
Регуляризация по технике Ridge (L2). Сохраняет все переменные и присваивает им важность на основании вклада в эффективность модели. Подходит для набора данных, включающего немного переменных, которые нужны для интерпретации полученных результатов.
Elastic-Net. Объединяет в себе преимущества обеих техник регуляризации. Реализуется посредством перекрестной валидации.

20. Что такое генеративно-состязательная сеть (GAN) и как она работает?
GAN — Generative adversarial Network — генеративно-состязательная сеть. Эта архитектура была придумана в 2014 году специально для задачи генерации новых изображений (конечно, ГАНы можно при этом использовать для генерации не только картинок, но и других объектов).
Модель состоит из двух отдельных нейросетей: генератора и дискриминатора. Генератор принимает на вход случайный вектор и выдает на выходе картинку. Дискриминатор принимает на вход картинку и выдает ответ: реальная ли эта картинка или фейковая (сгенерированная генератором).

Генератор и дискриминатор обучаются совместно. Дискриминатор помогает генератору научиться генерировать разнообразные картинки, а не только те, что есть в обучающем датасете.
Смысл в этом всем такой: дискриминатор быстро учится отличать картинки, сгенерированные генератором, от настоящих (из обучающего датасета). После этого на стадии обучения генератора генератор учится подстраиваться так, чтобы генерировать картинки, которые дискриминатор отличить от настоящих уже не сможет. После этого дискриминатор учится отличать новые картинки, сгенерированные генератором, от настоящих. После этого снова генератор учится генерировать картинки лучше. И так далее.
